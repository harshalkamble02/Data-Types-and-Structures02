 How It Works Internally:

Python dictionaries use a hash table under the hood.


When you create a key, Python calculates a hash value using the hash() function.

When you access a value using its key, Python quickly looks up the hash to find the value — making it very fast (O(1) time).

Values can be anything — even lists, other dictionaries, etc.

A hash table is a data structure that stores data using a key-value system. It uses a hashing function to convert keys into unique indexes, which point to the stored values. This allows for very fast data access — usually in constant time O(1).
 How It Relates to Dictionaries in Python:
In Python, the dictionary (dict) is implemented using a hash table.

Here’s what’s happening behind the scenes:

Q10 Explain why strings are immutable in Python?
- In Python, strings are immutable, which means once a string is created, it cannot be changed. Any operation that seems to modify a string actually creates a new string object.
Reasons Why Strings Are Immutable:

1. Performance and Memory Optimization
Python stores some strings in a shared string pool (called interning).

Since strings are immutable, they can be safely reused without copying or worrying about changes.
 2. Hashability for Dictionary Keys
Immutable types like strings can be hashed (converted into a fixed memory address).

This makes strings usable as keys in dictionaries and sets.
3. Thread-Safety and Predictability
In multi-threaded programs, immutable strings help prevent bugs caused by accidental changes.

Once created, the string’s value is safe to use anywhere without fear of being modified.

4. Simplicity and Security
It’s easier and safer to read and trust data that can't be changed mid-process.

Especially useful in secure systems, where modifying sensitive strings (like passwords) during execution could be risky.

Q17 Describe the time complexity of accessing elements in a dictionary.
-  In Python, accessing elements in a dictionary using a key has a time complexity of:
 Average Case:

O(1) — Constant Time
This means you can access any value instantly, regardless of how large the dictionary is.
Why Is It O(1)?
Python dictionaries use a hash table internally:

The key is passed to a hash function.

The hash value determines the index where the value is stored.

Python jumps directly to that index to fetch the value — no searching needed.

 -  Worst Case Time Complexity:
O(n)
This can happen rarely, if many keys hash to the same index (hash collisions) — then Python has to search through a group of keys at that spot. However, Python handles collisions efficiently, so this is uncommon.

Q19 Why are dictionaries considered unordered, and how does that affect data retrieval?
-  Why Dictionaries Were Considered Unordered (Before Python 3.7):

Traditionally, dictionaries were called unordered because the insertion order of keys was not guaranteed. This meant:

The way you inserted items was not the same as the way Python stored or returned them.

This behavior made dictionaries unpredictable in order, especially in older versions of Python.

 What Changed in Python 3.7+:

Starting from Python 3.7, dictionaries do preserve insertion order as an official feature of the language.

However, they are still considered unordered in concept, because:

The primary purpose of a dictionary is fast access by key, not maintaining order.

Python’s internal handling is based on hashing, not indexing.

 How Does That Affect Data Retrieval?
Access by Key, Not Position

You retrieve values using keys:

The dictionary doesn't sort keys for you — if you need sorted output, you must do it manually:

Even though insertion order is preserved now, you should not design logic that depends on it — the focus should always be on keys and values.
